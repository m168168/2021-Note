```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from subprocess import check_output
import seaborn as sns
from sklearn import datasets
```

1： 导入原始数据

```python
breast_cancer = datasets.load_breast_cancer()
# breast_cancer.target_names 
data = pd.DataFrame(data=breast_cancer.data,columns=breast_cancer.feature_names)
target = pd.DataFrame(data=breast_cancer.target,columns=['label'])
print(breast_cancer.target.shape)
```

2： 数据分析

```python
data.head()
col =data.columns
target['label'] = target['label'].apply( lambda x: 'good' if x==0 else 'bad')
#（variance）和偏差（bias）以及数据中最大值和最小值是多少。这些类型的信息有助于了解数据的状态，
data.describe() 
```

```python
# 分析各个数据的占比 根据标签 否则某个取值 
sns.countplot(target['label'] ,label='Count')
```

```python
y = pd.DataFrame(breast_cancer.target,columns=['label']) # 标签 
# 将数据标准化
data_n_2 = (data -data.mean()) /data.std()
#选取前十个数据
data_n_3 = pd.concat([y,data_n_2.iloc[:,0:10]],axis=1) # 前10个数据展示
data_n_3 = pd.melt(data_n_3,id_vars="label",var_name="features",value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="label", data=data_n_3,split=True, inner="quart")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x="features", y="value", hue="label", data=data_n_3)
plt.xticks(rotation=90)
```

```python
# 两个变量的分析
sns.jointplot(data.loc[:,'mean radius'], data.loc[:,'mean texture'], kind="reg", color="#ce1414")
```

```python
# 多个变量关系
sns.set(style="white")
df = data.loc[:,['worst radius','worst compactness','worst area']]
g = sns.PairGrid(df, diag_sharey=False)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw=3)
```

```python
# 数据的展示
sns.set(style="whitegrid", palette="muted")
# 将数据标准化
data_n_2 = (data -data.mean()) /data.std()
#选取前十个数据
data_n_3 = pd.concat([y,data_n_2.iloc[:,0:10]],axis=1)
data_n_3 = pd.melt(data_n_3,id_vars="label",var_name="features",value_name='value')
plt.figure(figsize=(10,10))
tic = time.time()
sns.swarmplot(x="features", y="value", hue="label", data=data_n_3)
plt.xticks(rotation=90)
```

```python
# 多个变量的分析  热力图
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
#分布图 某一列数据的分布图 
sns.distplot(data['mean radius'])
```
